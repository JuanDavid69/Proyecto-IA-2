### Cargamos todos los datos en un dataFrame

Datos obtenidos de https://www.kaggle.com/ronitf/heart-disease-uci

Esta base de datos contiene 76 atributos, pero todos los experimentos publicados se refieren al uso de un subconjunto de 14 de ellos. El campo "target" se refiere a la presencia de enfermedad cardíaca en el paciente. Tiene un valor entero de 0 (sin presencia) a 1.

**Describcion de las caracteristicas:**
- **age** Edad en años 
- **sex** 1 = hombre; 0 = mujer
- **cp** Tipo de dolor en el pecho
- **trestbps** Presión arterial en reposo (en mm Hg al ingreso al hospital)
- **chol** Suero colestoral en mg / dl
- **fbs** Azúcar en sangre en ayunas > 120 mg / dl (1 = True; 0 = False)
- **restecg** Resultados electrocardiográficos en reposo (0 = normal; 1 = anormalidad de la onda ST-T; 2 = hipertrofia ventricular izquierda probable o definitiva)
- **thalach** Frecuencia cardíaca máxima alcanzada
- **exang** Angina inducida por ejercicio (1 = sí; 0 = no)
- **oldpeak** Depresión del ST inducida por el ejercicio relativo al descanso
- **slope** La pendiente del segmento ST de ejercicio pico (0 = ascenso; 1 = plano; 2 = descenso)
- **ca** Número de vasos principales coloreados por flourosopía
- **thal** 3 = normal; 6 = defecto fijo; 7 = defecto reversible
- **target** 1(presencia de enfermedad cardíaca) o 0 (sin presencia)

Antes de continuar con la implementación de los modelos, se debe asegurar que no existen valores vacíos dentro del conjunto de datos ya que esto puede generar inconvenientes durante el entrenamiento.

```python
print(heart.isnull().sum().sum())
```

### Dividimos el conjunto de datos en 80% para entrenamiento y 20% para pruebas

```python
from sklearn.model_selection import train_test_split

variables = heart[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope','ca','thal']].values
target = heart['target'].values

trainX, testX, trainY, testY = train_test_split(variables, target, test_size=0.2)
print("Datos para entrenamiento:")
print(trainX.shape, trainY.shape ,'\n')
print("Datos para pruebas:")
print(testX.shape, testY.shape)
```

## Árbol de decisión

Es un diagrama de construcción lógica que representa y categoriza una serie de condiciones para falicitar la resolución de un objetivo claro. El árbol se encuentra compuesto por nodos que representan el momento donde se debe tomar un decisión y las hojas que representan el resultado final de la decisión tomada.

#### Definición de parametros:

**max_depth:** int o None, opcional (predeterminado = Ninguno)
- La profundidad máxima del árbol. Si ninguno, los nodos se expanden hasta que todas las hojas sean puras o hasta que todas las hojas contengan menos de min_samples_split samples.

**min_samples_split:** int, float, opcional (predeterminado = 2)
El número mínimo de muestras necesarias para dividir un nodo interno:

- Si int, entonces considere **min_samples_split** como el número mínimo.
- Si flota, **min_samples_split** es una fracción y ceil (min_samples_split * n_samples) es el número mínimo de muestras para cada división.

**random_state:** int, instancia RandomState o None, opcional (predeterminado = Ninguno)

- Si int, **random_state** es la semilla utilizada por el generador de números aleatorios; Si la instancia de RandomState, **random_state** es el generador de números aleatorios; Si no hay ninguno, el generador de números aleatorios es la instancia RandomState utilizada por np.random.

```python
max_depth = 10
min_samples_split = 5
myTree = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 5, random_state = 1)
myTree = myTree.fit(trainX, trainY)

print('Puntaje en los datos de entrenamiento:')
print(myTree.score(trainX, trainY), '\n')
print('Puntaje en los datos de pruebas:')
print(myTree.score(testX, testY))
```

### Matriz de confusión para el modelo de Árbol de decisión 

```python
from sklearn.metrics import confusion_matrix
predicted = myTree.predict(testX)
expected = testY
cm1 = confusion_matrix(expected, predicted)
print(cm1)
print("'NOT HEART DISEASE' clasificados como 'NOT HEART DISEASE':" , cm1[0,0])
print("'HEART DISEASE' clasificados como 'NOT HEART DISEASE': ", cm1[0,1])
print("'NOT HEART DISEASE' clasificados como 'HEART DISEASE':" , cm1[1,0])
print("'HEART DISEASE' clasificados como 'HEART DISEASE':" , cm1[1,1])
```

### Interpretación del modelo obtenido

```python
feature_name =['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope','ca','thal']
class_name = ["thersn't heart disease",'heart disease']
dot_data = tree.export_graphviz(myTree,filled=True,feature_names=feature_name,class_names=class_name , rounded=True, special_characters=True, )  
graph = graphviz.Source(dot_data)  
graph 
```

## Modelo Naive Bayes
Un modelo de Naive Bayes o modelo de Bayes ingenuo es un clasificador basado en el teorema de Bayes y otras hipotésis de simplificación adicionales. Es gracias a estas simplificaciones que se le da el nombre de ingenuo, ya que se suelen resumir en la hipótesis de independencia entre las variables predictoras.

Generalmente se utilizan 3 tipos de clasificadores, Gaussiano, Bernoulli y Multinomial. Los dos últimos son utilizados comúnmente para características discretas como en filtrado de documentos o spam. El clasificador Gaussiano tiene una distribución normal y es usado para características con valores continuos.

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

myGaussianNB = gnb.fit(trainX, trainY)

print('Puntaje en los datos de entrenamiento:')
print(myGaussianNB.score(trainX, trainY), '\n')
print('Puntaje en los datos de pruebas:')
print(myGaussianNB.score(testX, testY))
```

La función fit() recibe dos parámetros fundamentales, en este caso trainX y trainY. El primero representa las muestras y el segundo parámetro los valores objetivos. De este forma, el modelo se entrena con los valores iniciales dados para después ser probado y comparado con los valores de pruebas.

### Matriz de confusión para el modelo de  Naive Bayes

```python
cm2 = confusion_matrix(testY, myGaussianNB.predict(testX))
print(cm2)
print("'NOT HEART DISEASE' clasificados como 'NOT HEART DISEASE':" , cm2[0,0])
print("'HEART DISEASE' clasificados como 'NOT HEART DISEASE': ", cm2[0,1])
print("'NOT HEART DISEASE' clasificados como 'HEART DISEASE':" , cm2[1,0])
print("'HEART DISEASE' clasificados como 'HEART DISEASE':" , cm2[1,1])
```

### Interpretación del modelo obtenido

Para la mayoría de las pruebas ejecutadas, la matriz de confusión indica un nivel de precisión bastante acertado utilizando el modelo de Naive Bayes. Incluso, es más preciso que el árbol de decisión utilizado anteriormente para calcular los valores de la columna target.

## Neural network
#### Definición de parametros:

**solver:** El solucionador para la optimización del peso.

- "lbfgs" es un optimizador en la familia de los métodos cuasi-Newton.
- "sgd" se refiere al descenso de gradiente estocástico.
- "adam" se refiere a un optimizador estocástico basado en gradiente propuesto por Kingma, Diederik y Jimmy Ba 

**alpha:** Parámetro de penalización L2 (término de regularización). 

**activation:** Función de activación para la capa oculta.

- "identity", activación no operativa, útil para implementar un cuello de botella lineal, devuelve *f (x) = x*
- "logistic", la función sigmoide logística, devuelve *f (x) = 1 / (1 + exp (-x))*
- "tanh", la función hiperbólica de tan, devuelve *f (x) = tanh (x)*
- "relu", la función de unidad lineal rectificada, devuelve *f (x) = max (0, x)*

**max_iter:** Número máximo de iteraciones. El solucionador itera hasta la convergencia o este número de iteraciones. 

**hidden_layer_sizes:** El elemento i-ésimo representa el número de neuronas en la i-ésima capa oculta.

**random_state:** Si es int, **random_state** es la semilla utilizada por el generador de números aleatorios; Si es instancia de RandomState, **random_state** es el generador de números aleatorios; Si no hay ninguno, el generador de números aleatorios es la instancia RandomState utilizada por np.random.

**verbose:** Si se deben imprimir mensajes de progreso en stdout.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  
trainXX = trainX.copy()
testXX = testX.copy()
scaler.fit(trainX)  
trainXX = scaler.transform(trainXX)
testXX = scaler.transform(testXX)  

myNN = MLPClassifier(solver='lbfgs', alpha=1e-3, activation = 'relu', max_iter=5000,
                     hidden_layer_sizes = (1), random_state=1, verbose = True)

myNN.fit(trainXX, trainY)                         

print('Puntaje en los datos de entrenamiento:')
print(myNN.score(trainXX, trainY), '\n')
print('Puntaje en los datos de pruebas:')
print(myNN.score(testXX, testY))
```

### Matriz de confusión para la Neural network

```python
predicted3 = myNN.predict(testXX)
expected3 = testY
cm3 = confusion_matrix(expected3, predicted3)
print(cm3)
print("'NOT HEART DISEASE' clasificados como 'NOT HEART DISEASE':" , cm3[0,0])
print("'HEART DISEASE' clasificados como 'NOT HEART DISEASE': ", cm3[0,1])
print("'NOT HEART DISEASE' clasificados como 'HEART DISEASE':" , cm3[1,0])
print("'HEART DISEASE' clasificados como 'HEART DISEASE':" , cm3[1,1])
```

### Interpretación del modelo obtenido
