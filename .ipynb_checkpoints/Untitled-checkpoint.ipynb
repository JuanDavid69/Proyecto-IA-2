```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
heart = pd.read_csv('heart.csv')
heart.head()
```

```python
fig, axs =plt.subplots(2, 7, figsize =(20,7))

axs[0,0].hist(heart.age, label='Edad')

axs[0,1] = heart.hist(column='trestbps')
plt.title('Resting blood pressure')

axs[0,2] = heart.hist(column='chol')
plt.title('Serum cholestoral in mg/dl')

axs[0,3] = heart.hist(column='thalach')
plt.title('Maximum heart rate achieved')

axs[0,5] = heart.hist(column='oldpeak')
plt.title('ST depression induced by exercise relative to rest')

plt.show()
```
